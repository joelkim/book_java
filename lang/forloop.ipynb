# for 루프

## 일반적 for 루프

- 카운터 변수 사용

```java
for (카운터 변수 초기화; 카운터 변수 조건문; 카운터 변수 변경) {
  statements
}
```

## 강화 for 루프

- 콜렉션 객체에 대해 사용 가능

```java
for (콜랙션 개체의 요소 : 콜랙션 객체) {
  statements
}
```

- 배열 콜렉션의 경우

````java
int[] array = {1, 2, 3};
for (int number : array) {
   System.out.printf("%d%n", number);
}
````
